```python
import json 
```

```python
with open("public/stress-performance/assets/data/orders.json", 'r') as fdata:
    data = json.load(fdata)
del data['training']
```

```python
Config['components'] = [dict() for i in range(9) ]
```

```python
sizes = Config['components']
```

```python
for component in sizes: 
    component['order'] = 'latinSquare'
    component['components'] = [dict() for i in range(3)]
    for stimuli in component['components']:
        stimuli['order'] = 'fixed'
        stimuli['components'] = ['string' for i in range(9)]
```

```python
for group, gcomponent in zip(data.keys(), Config['components']):
    print('group', group)
    gcomponent['interruptions'] = [
                            {
                                "spacing": 1,
                                "firstLocation": 1,
                                "components": [
                                    "break"
                                ]
                            }
                        ]
    for size, scomponent in zip(data[group].keys(), gcomponent['components']):
        print('size', size)
        scomponent['components'] = {
            "fake1",
            "fake2",
            {
            "order": "random", 
            "components": [f"spl_{group}_{size}_{i}" for i in range(9)]
            }
        }
```

```python
"""
"shortest-path-1":{
    "baseComponent": "shortest-path",
    "description": "A trial for the user to click the largest bar",
    "instruction": "Click on the largest bar",
    "parameters": {
        "indices": ["1", "10", 0]
    }
},       
"""
componentdefs = dict()
for group in range(1,10):
    for size in [10,25,50]:
        for index in range(9):
            componentdefs[f'spl_{group}_{size}_{index}'] = {
                "baseComponent": "shortest-path",
                "description": "Shortest path stimulus",
                "instruction": "What is the length of the shortest path between the highlighted nodes?",
                "parameters": {
                    "indices": [group, size, index]
                }
            }
```
